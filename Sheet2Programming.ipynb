# 3.1 
### Write a Matlab / Python function which plots f(t) and its magnitude spectrum.
### f(t) := cos(4πt) + 4 cos(10πt) + 8 cos(2π20t).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 512
Fs = 1/N
t = np.arange(0, 10, 10/N) #t ∈ {k/N | k ∈ [0 : N − 1]}
print(len(t))
print(t[511])
```

```python
func = np.cos(4 * np.pi * t) + 4*np.cos(10 * np.pi * t) + 8* np.cos(2 * np.pi*20* t)
```

```python
fig, axs = plt.subplots()

axs.set_title("Magnitude Spectrum")
axs.magnitude_spectrum(func, Fs=Fs, color='C1')

plt.show()
```

## 3.3 
### Write a Matlab / Python function that plots the Nth-roots of unity and the primitive Nth-roots of unity for a given N. 
### Test your function for N = 6.


```python
import numpy as np
import matplotlib.pyplot as plt

def plot_roots_of_unity(N):
    omega_N = np.exp(2 * np.pi * 1j / N)
    roots_of_unity = [omega_N ** k for k in range(N)]
    primitive_roots_of_unity = [omega_N ** k for k in range(N) if np.gcd(k, N) == 1] # ρ := exp(2πi/N) is a primitive N-th root of unity, i.e., ρm = 1 if m ≡ 0 mod N
    t = np.linspace(0,np.pi*2,100)
    plt.figure(figsize=(6, 6))
    plt.plot(np.cos(t), np.sin(t), color = 'black', label = "Unit Circle")
    plt.scatter(np.real(roots_of_unity), np.imag(roots_of_unity), label="Roots of Unity")
    plt.scatter(np.real(primitive_roots_of_unity), np.imag(primitive_roots_of_unity), label="Primitive Roots of Unity")
    plt.xlim((-1.5, 1.5))
    plt.ylim((-1.5, 1.5))
    plt.ylabel('Imaginary')
    plt.xlabel('Real')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.legend()
    plt.show()
```

```python
plot_roots_of_unity(6)
```
