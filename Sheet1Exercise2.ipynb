```python
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

### a - Create a 5 seconds long discrete sine wave of frequency 1000 Hz.The sampling rate is 44100 Hz.


```python
fs = 44100 #sampling freqency
f = 1000 #frequency of the sine wave
t = 5 #seconds

samples = np.arange(t * fs) / fs
sine_signal = np.sin(2 * np.pi * f * samples).astype(np.float32)
```

### b - Plot the spectrogram of the signal you created in (a)

```python
# Plot the signal
plt.subplot(211)
plt.title('1000Hz sine wave')
plt.plot(samples,sine_signal)
plt.xlabel('Sample')
plt.ylabel('Amplitude')
# Plot the spectrogram
plt.subplot(212)
powerSpectrum, freqenciesFound, time, imageAxis = plt.specgram(sine_signal, Fs=fs)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Spectrogram of a 1000Hz sine wave')
plt.tight_layout()
plt.show()   
```

On the plot depicting sine wive, we cannot really see a clear sine wave - instead it looks like a rectangle. It is because the frequency of the wave is so high, we can't see it on the plot in the span of 5 seconds. Nevertheless, the spectrogram of the sine wave is very characteristic and we can see it very clearly on the plot below. It is a straight line at the 1000Hz frequency (frequency of the sine wave).  

### c - Repeat (a) and (b) with a cosine wave of frequency 1000 Hz. Compare the two spectrograms

```python
cosine_signal = np.cos(2 * np.pi * f * samples).astype(np.float32)

plt.subplot(211)
plt.title('1000Hz cosine wave')
plt.plot(samples,cosine_signal)
plt.xlabel('Sample')
plt.ylabel('Amplitude')
# Plot the spectrogram
plt.subplot(212)
powerSpectrum, freqenciesFound, time, imageAxis = plt.specgram(cosine_signal, Fs=fs)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Spectrogram of a 1000Hz cosine wave')
plt.tight_layout()
plt.show()
```

### Comments: There doesn't seem to be a difference between sine and cosine spectrograms. This is because cosine wave is simply a phase-shifted version of a sine wave. The fourier transform of both signals is the same except from phase shift in freqency domain. Nevertheless spectrograms of both these signals contain the same information about freqency content of the signal, so in the end **spectrograms are the same**. 

### Save the signal you created in (a) in a .wav file.

```python
wavfile.write(f"sinewave{f}Hz.wav", fs, sine_signal)
```
