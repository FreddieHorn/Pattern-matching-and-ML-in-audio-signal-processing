```python
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

```python
fs = 44100 #sampling freqency
f = 100 #frequency of the sine wave
t = 5 #seconds

samples = np.arange(t * fs) / fs
signal = np.sin(2 * np.pi * f * samples)
plt.title('Spectrogram of a 1000Hz sine wave')
# Plot the signal
plt.subplot(211)
plt.plot(samples,signal)
plt.xlabel('Sample')
plt.ylabel('Amplitude')
# Plot the spectrogram
plt.subplot(212)
powerSpectrum, freqenciesFound, time, imageAxis = plt.specgram(signal, Fs=fs)
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()   
```
